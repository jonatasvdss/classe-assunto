```python
import polars as pl

caminho_csv = r'..\..\data\v2_ftp\raw\assuntos_recorrentes.csv'
df_assuntos = pl.read_csv(caminho_csv, separator='#')

print(f"Total de linhas e colunas: {df_assuntos.shape}")
display(df_assuntos.head())
```

```python
# Me foi informado que talvez a coluna de assuntos pudesse
# conter múltiplos assuntos separados por um "@"
df_assuntos.filter(pl.col('assunto').str.contains('@'))
```

```python
nulos = df_assuntos.null_count()
print("Valores Nulos por Coluna:")
display(nulos)

duplicados = df_assuntos.is_duplicated().sum()
print(f"\nLinhas duplicadas no dataset: {duplicados}")
```

```python
distribuicao_assuntos = (
    df_assuntos.group_by("assunto")
    .len(name="quantidade")
    .sort("quantidade", descending=True)
)

for i, (assunto, _) in enumerate(distribuicao_assuntos.iter_rows(), start=1):
    print(f'{i}: {assunto}')
```

# **Dataset Processado**

```python
import polars as pl

caminho_csv = r'..\..\data\v2_ftp\processed\assuntos_recorrentes_processado.csv'
df_assuntos_processado = pl.read_csv(caminho_csv)

print(f"Total de linhas e colunas: {df_assuntos_processado.shape}")
display(df_assuntos_processado.head())
```

```python
distribuicao_assuntos_processado = (
    df_assuntos_processado.group_by("assunto")
    .len(name="quantidade")
    .sort("quantidade", descending=True)
)

nomes_assuntos = []
for assunto, _ in distribuicao_assuntos_processado.iter_rows():
    nomes_assuntos.append(assunto)

for i, a in enumerate(sorted(nomes_assuntos), start=1):
    print(f'{i}: {a}')
```
