```python
import polars as pl

caminho_csv = r'..\..\data\v2_ftp\raw\classes_recorrentes.csv'

df_classes = pl.read_csv(caminho_csv, separator='#')

print(f"Total de linhas e colunas: {df_classes.shape}")
display(df_classes.head())
```

```python
nulos = df_classes.null_count()
print("Valores Nulos por Coluna:")
display(nulos)

duplicados = df_classes.is_duplicated().sum()
print(f"\nLinhas duplicadas no dataset: {duplicados}")

if duplicados > 0:
    df_classes = df_classes.unique(subset=["numero_processo"], keep="first")
    print(f"Total de linhas após remover duplicatas: {df_classes.shape[0]}")
```

```python
distribuicao_classes = (
    df_classes.group_by("classe")
    .len(name="quantidade")
    .sort("quantidade", descending=True)
)

print(f"Total de classes únicas: {distribuicao_classes.shape[0]}")
display(distribuicao_classes.tail(10))

for i, (classe, _) in enumerate(distribuicao_classes.iter_rows(), start=1):
    print(f'{i}: {classe}')
```

# **Dataset Processado**

```python
import polars as pl

caminho_csv_processado = r'..\..\data\v2_ftp\processed\classes_recorrentes_processado.csv'

df_classes_processado = pl.read_csv(caminho_csv_processado)

print(f"Total de linhas e colunas: {df_classes_processado.shape}")
display(df_classes_processado.head())
```

```python
distribuicao_classes_processado = (
    df_classes_processado.group_by("classe")
    .len(name="quantidade")
    .sort("quantidade", descending=True)
)

print(f"Total de classes únicas: {distribuicao_classes_processado.shape[0]}")
display(distribuicao_classes_processado.tail(10))

nomes_classes = []
for classe, _ in distribuicao_classes_processado.iter_rows():
    nomes_classes.append(classe)

for i, c in enumerate(sorted(nomes_classes), start=1):
    print(f'{i}: {c}')
```
