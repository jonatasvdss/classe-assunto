```python
import sys
import os
import joblib
import pandas as pd
import polars as pl
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

sys.path.append(os.path.abspath('../..'))

from src.train_models import transformar_dados_teste
from src.evaluate import avaliar_modelo

caminho = r'..\..\data\v2_ftp\processed\classes_recorrentes_processado.csv'
df = pl.read_csv(caminho)
```

```python
caminho_vet_classe = r'..\..\models\v2_ftp\vectors\vetorizador_tfidf_classe.joblib'

X_teste_classe, y_teste_classe = transformar_dados_teste(
    df,
    caminho_vetorizador=caminho_vet_classe,
    coluna_texto="texto_limpo",
    coluna_alvo="classe"
)

resultados_modelos_classe = []

def registrar_metricas_classe(nome_modelo, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    _, _, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)
    resultados_modelos_classe.append({
        "Modelo": nome_modelo,
        "Acurácia": round(acc, 4),
        "F1-Score": round(f1, 4)
    })
```

```python
modelo_nb_classe = joblib.load(r'..\..\models\v2_ftp\modelo_naive_bayes_classe.joblib')

y_pred_nb_classe = modelo_nb_classe.predict(X_teste_classe)

avaliar_modelo(modelo_nb_classe, X_teste_classe, y_teste_classe, "Naive Bayes Classe", versao="v2_ftp")
registrar_metricas_classe("Naive Bayes", y_teste_classe, y_pred_nb_classe)
```

```python
modelo_rf_classe = joblib.load(r'..\..\models\v2_ftp\modelo_random_forest_classe.joblib')

y_pred_rf_classe = modelo_rf_classe.predict(X_teste_classe)

avaliar_modelo(modelo_rf_classe, X_teste_classe, y_teste_classe, "Random Forest Classe", versao="v2_ftp")
registrar_metricas_classe("Random Forest", y_teste_classe, y_pred_rf_classe)
```

```python
modelo_lr_classe = joblib.load(r'..\..\models\v2_ftp\modelo_regressao_logistica_classe.joblib')

y_pred_lr_classe = modelo_lr_classe.predict(X_teste_classe)

avaliar_modelo(modelo_lr_classe, X_teste_classe, y_teste_classe, "Regressao Logistica Classe", versao="v2_ftp")
registrar_metricas_classe("Regressão Logística", y_teste_classe, y_pred_lr_classe)
```

```python
le_classe = joblib.load(r'..\..\models\v2_ftp\labels\label_encoder_classe.joblib')
y_teste_classe_enc = le_classe.transform(y_teste_classe)
```

```python
modelo_lgbm_classe = joblib.load(r'..\..\models\v2_ftp\modelo_lightgbm_classe.joblib')
y_pred_lgbm_classe_enc = modelo_lgbm_classe.predict(X_teste_classe)
y_pred_lgbm_classe = le_classe.inverse_transform(y_pred_lgbm_classe_enc)

avaliar_modelo(modelo_lgbm_classe, X_teste_classe, y_teste_classe_enc, "LightGBM Classe", versao="v2_ftp")
registrar_metricas_classe("LightGBM", y_teste_classe, y_pred_lgbm_classe)
```

```python
modelo_xgb_classe = joblib.load(r'..\..\models\v2_ftp\modelo_xgboost_classe.joblib')
y_pred_xgb_classe_enc = modelo_xgb_classe.predict(X_teste_classe)
y_pred_xgb_classe = le_classe.inverse_transform(y_pred_xgb_classe_enc)

avaliar_modelo(modelo_xgb_classe, X_teste_classe, y_teste_classe_enc, "XGBoost Classe", versao="v2_ftp")
registrar_metricas_classe("XGBoost", y_teste_classe, y_pred_xgb_classe)
```
