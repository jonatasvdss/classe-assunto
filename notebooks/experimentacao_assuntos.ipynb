```python
import sys
import os
import time
import pandas as pd
import polars as pl
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

sys.path.append(os.path.abspath('..'))

from src.train_models import extrair_features_e_dividir, salvar_modelo_e_vetorizador
from src.evaluate import avaliar_modelo

caminho = r'..\data\processed\dataset_assuntos_processado.csv'
df = pl.read_csv(caminho)

print(f"Dataset carregado e limpo. Formato final: {df.shape}")
```

```python
X_treino_assunto, X_teste_assunto, y_treino_assunto, y_teste_assunto, vetorizador_assunto = extrair_features_e_dividir(
    df, 
    coluna_texto="texto_limpo", 
    coluna_alvo="assunto"
)

resultados_modelos_assunto = []

def registrar_metricas_assunto(nome_modelo, y_true, y_pred, tempo_treino):
    acc = accuracy_score(y_true, y_pred)
    _, _, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)
    resultados_modelos_assunto.append({
        "Modelo": nome_modelo,
        "Acurácia": round(acc, 4),
        "F1-Score": round(f1, 4),
        "Tempo (s)": round(tempo_treino, 2)
    })
```

```python
inicio_nb_assunto = time.time()
modelo_nb_assunto = MultinomialNB()
modelo_nb_assunto.fit(X_treino_assunto, y_treino_assunto)
tempo_nb_assunto = time.time() - inicio_nb_assunto

y_pred_nb_assunto = modelo_nb_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_nb_assunto, X_teste_assunto, y_teste_assunto, "Naive Bayes Assunto")
salvar_modelo_e_vetorizador(
    modelo_nb_assunto, 
    vetorizador_assunto, 
    "modelo_naive_bayes_assunto", 
    "vetorizador_tfidf_assunto"
)
registrar_metricas_assunto("Naive Bayes", y_teste_assunto, y_pred_nb_assunto, tempo_nb_assunto)
```

```python
inicio_rf_assunto = time.time()
modelo_rf_assunto = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
modelo_rf_assunto.fit(X_treino_assunto, y_treino_assunto)
tempo_rf_assunto = time.time() - inicio_rf_assunto

y_pred_rf_assunto = modelo_rf_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_rf_assunto, X_teste_assunto, y_teste_assunto, "Random Forest Assunto")
salvar_modelo_e_vetorizador(
    modelo_rf_assunto, 
    vetorizador_assunto, 
    "modelo_random_forest_assunto", 
    "vetorizador_tfidf_assunto"
)
registrar_metricas_assunto("Random Forest", y_teste_assunto, y_pred_rf_assunto, tempo_rf_assunto)
```

```python
inicio_lr_assunto = time.time()
modelo_lr_assunto = LogisticRegression(
    max_iter=1000, 
    n_jobs=-1, 
    random_state=42, 
    class_weight='balanced',
    C=2.0
)
modelo_lr_assunto.fit(X_treino_assunto, y_treino_assunto)
tempo_lr_assunto = time.time() - inicio_lr_assunto

y_pred_lr_assunto = modelo_lr_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_lr_assunto, X_teste_assunto, y_teste_assunto, "Regressao Logistica Assunto")
salvar_modelo_e_vetorizador(
    modelo_lr_assunto, 
    vetorizador_assunto, 
    "modelo_regressao_logistica_assunto", 
    "vetorizador_tfidf_assunto"
)
registrar_metricas_assunto("Regressão Logística", y_teste_assunto, y_pred_lr_assunto, tempo_lr_assunto)
```

```python
df_comparacao_assunto = pd.DataFrame(resultados_modelos_assunto)
display(df_comparacao_assunto.sort_values(by="F1-Score", ascending=False))
```
