```python
import sys
import os
import joblib
import pandas as pd
import polars as pl
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

sys.path.append(os.path.abspath('..'))

from src.train_models import transformar_dados_teste
from src.evaluate import avaliar_modelo

caminho = r'..\data\v1_original\processed\assuntos_recorrentes_processado.csv'
df = pl.read_csv(caminho)
```

```python
caminho_vet_assunto = r'..\models\v1_original\vectors\vetorizador_tfidf_assunto.joblib'

X_teste_assunto, y_teste_assunto = transformar_dados_teste(
    df,
    caminho_vetorizador=caminho_vet_assunto,
    coluna_texto="texto_limpo",
    coluna_alvo="assunto"
)

resultados_modelos_assunto = []

def registrar_metricas_assunto(nome_modelo, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    _, _, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)
    resultados_modelos_assunto.append({
        "Modelo": nome_modelo,
        "Acurácia": round(acc, 4),
        "F1-Score": round(f1, 4)
    })
```

```python
modelo_nb_assunto = joblib.load(r'..\models\v1_original\modelo_naive_bayes_assunto.joblib')

y_pred_nb_assunto = modelo_nb_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_nb_assunto, X_teste_assunto, y_teste_assunto, "Naive Bayes Assunto")
registrar_metricas_assunto("Naive Bayes", y_teste_assunto, y_pred_nb_assunto)
```

```python
modelo_rf_assunto = joblib.load(r'..\models\v1_original\modelo_random_forest_assunto.joblib')

y_pred_rf_assunto = modelo_rf_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_rf_assunto, X_teste_assunto, y_teste_assunto, "Random Forest Assunto")
registrar_metricas_assunto("Random Forest", y_teste_assunto, y_pred_rf_assunto)
```

```python
modelo_lr_assunto = joblib.load(r'..\models\v1_original\modelo_regressao_logistica_assunto.joblib')

y_pred_lr_assunto = modelo_lr_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_lr_assunto, X_teste_assunto, y_teste_assunto, "Regressao Logistica Assunto")
registrar_metricas_assunto("Regressão Logística", y_teste_assunto, y_pred_lr_assunto)
```

```python
modelo_svc_assunto = joblib.load(r'..\models\v1_original\modelo_linearsvc_assunto.joblib')

y_pred_svc_assunto = modelo_svc_assunto.predict(X_teste_assunto)

avaliar_modelo(modelo_svc_assunto, X_teste_assunto, y_teste_assunto, "LinearSVC Assunto")
registrar_metricas_assunto("LinearSVC", y_teste_assunto, y_pred_svc_assunto)
```

```python
le_assunto = joblib.load(r'..\models\v1_original\labels\label_encoder_assunto.joblib')
y_teste_assunto_enc = le_assunto.transform(y_teste_assunto)
```

```python
modelo_lgbm_assunto = joblib.load(r'..\models\v1_original\modelo_lightgbm_assunto.joblib')
y_pred_lgbm_assunto_enc = modelo_lgbm_assunto.predict(X_teste_assunto)
y_pred_lgbm_assunto = le_assunto.inverse_transform(y_pred_lgbm_assunto_enc)

avaliar_modelo(modelo_lgbm_assunto, X_teste_assunto, y_teste_assunto_enc, "LightGBM Assunto")
registrar_metricas_assunto("LightGBM", y_teste_assunto, y_pred_lgbm_assunto)
```

```python
modelo_xgb_assunto = joblib.load(r'..\models\v1_original\modelo_xgboost_assunto.joblib')
y_pred_xgb_assunto_enc = modelo_xgb_assunto.predict(X_teste_assunto)
y_pred_xgb_assunto = le_assunto.inverse_transform(y_pred_xgb_assunto_enc)

avaliar_modelo(modelo_xgb_assunto, X_teste_assunto, y_teste_assunto_enc, "XGBoost Assunto")
registrar_metricas_assunto("XGBoost", y_teste_assunto, y_pred_xgb_assunto)
```
