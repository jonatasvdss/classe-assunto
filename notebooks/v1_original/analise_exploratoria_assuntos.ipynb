```python
import polars as pl
import seaborn as sns

sns.set_theme(style="whitegrid")

caminho_csv = r'..\data\v1_original\raw\assuntos_recorrentes.csv'
df_assuntos = pl.read_csv(caminho_csv, separator='#')

print(f"Total de linhas e colunas: {df_assuntos.shape}")
display(df_assuntos.head())
```

```python
nulos = df_assuntos.null_count()
print("Valores Nulos por Coluna:")
display(nulos)

duplicados = df_assuntos.is_duplicated().sum()
print(f"\nLinhas duplicadas no dataset: {duplicados}")

if duplicados > 0:
    df_assuntos = df_assuntos.unique(subset=["numero_processo"], keep="first")
    print(f"Total de linhas após remover duplicatas: {df_assuntos.shape[0]}")
```

```python
distribuicao_assuntos = (
    df_assuntos.group_by("assunto")
    .len(name="quantidade")
    .sort("quantidade", descending=True)
)

print(f"Total de assuntos únicos: {distribuicao_assuntos.shape[0]}")
display(distribuicao_assuntos.tail(10))

assuntos_unicos = distribuicao_assuntos.filter(pl.col("quantidade") == 1).shape[0]
print(f"\nAssuntos que aparecem apenas 1 vez: {assuntos_unicos}")
```
